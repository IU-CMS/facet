```python
import ROOT
import uproot
import numpy as np
import tqdm
```

```python
aD_p4 = ROOT.TLorentzVector(0., 0., 0., 0.)

file = ROOT.TFile.Open("data_Sibyll23c_100kEvnt_14TeVpp__J_Psy.root")

treeReader = ROOT.TTreeReader("tree", file)
jpsi_p4 = ROOT.TTreeReaderValue(ROOT.TLorentzVector)(treeReader, "J_Psy_p4")
```

```python
n_entries = treeReader.GetEntries(1)
print(n_entries)
```

```python
ybins  = np.linspace(5, 8, 31)
pTbins = np.linspace(2., 10, 17)
line = []
```

```python
for i_y in range(len(ybins)-1):
    for i_pt in range(len(pTbins)-1):
        xsec = 0
        for i_event in range(int(n_entries)):
            treeReader.SetEntry(i_event)
            if jpsi_p4.M()>0:
                y  = jpsi_p4.Rapidity()
                pt = jpsi_p4.Pt()
                if (y > ybins[i_y] and  y < ybins[i_y+1]):
                    if (pt > pTbins[i_pt] and  pt < pTbins[i_pt+1]):
                        xsec +=1
        print(ybins[i_y], ybins[i_y+1], pTbins[i_pt], pTbins[i_pt+1], xsec)
```
